### Distributing amongs the age groups according to national age trend

```python
def calculate_number_at_each_specific_age_in_range(age_min , age_max, reference_df):
    # First calculate the total number in the reference df age group
    total_range_reference_sum = 0
    for age in range(age_min , age_max+1):
        total_range_reference_sum += reference_df.loc[age,'VALUE']

    # Now return a list of total numbers at each age in the age group
    ratio_by_age = []
    for age in range(age_min , age_max+1):
        ratio_by_age += [reference_df.loc[age,'VALUE'] / total_range_reference_sum]
    return ratio_by_age
```

```python
mydf = df.loc[:,[str(x) for x in range(20)]]
mydf
```

```python
for key in ['20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80-84']:

    print('Working on age range = ',key)
    
    age_min, age_max = key.split('-')
    ratio_by_age = calculate_number_at_each_specific_age_in_range(int(age_min), int(age_max), national_age_curve)
    age_range_list = [str(a) for a in range(int(age_min), int(age_max)+1)]

    # The ages of the individuals
    
    for age in age_range_list:
        mydf.loc[:,age] = 0
    
    for index, row in mydf.iterrows():
        actual_ages_list = choice(age_range_list, df.loc[index, key], ratio_by_age)
        for age in actual_ages_list:
            mydf.loc[index,age] += 1
        
```
