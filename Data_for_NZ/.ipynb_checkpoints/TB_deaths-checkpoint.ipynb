```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_dummy = pd.read_csv('Age_and_sex.txt',sep='\t', header=2, thousands=',')
df_age = df_dummy.drop(labels=['#Males.','Females.','Males.', 'Females..1', 'Males..1', 'Females..2'], axis=1).drop(labels=['Unspecified minors', '80 and over', 'Unspecified adults'], axis=0).reset_index().rename(columns={'Males..2' : 'Males', 'Females..3' : 'Females', 'index':'Ages'})
df_age
```

```python
df_dummy = pd.read_csv('TB_age_of_persons_died_1916.txt',sep='\t', header=1)
df_TB_1916 = df_dummy.rename(columns={'#Ages.' : 'Ages', 'Males.' : 'Males 1916', 'Females.':'Females 1916'}).drop('Total.', axis=1).drop([16,17], axis=0)
df_TB_1916
```

```python
df_TB_1916['Male Deaths per 10,000 1916'] = df_TB_1916.loc[:,'Males 1916'].astype('float') / df_age.loc[:,'Males'].astype('float') * 10000
df_TB_1916
```

```python
df_dummy = pd.read_csv('TB_age_of_persons_died_1918.txt',sep='\t', header=1)
df_TB_1918 = df_dummy.rename(columns={'#Ages.' : 'Ages', 'Males.' : 'Males 1918', 'Females.':'Females 1918'}).drop('Total.', axis=1).drop([16,17], axis=0)
df_TB_1918
```

```python
df_TB_1918['Male Deaths per 10,000 1918'] = df_TB_1918.loc[:,'Males 1918'].astype('float') / df_age.loc[:,'Males'].astype('float') * 10000
df_TB_1918
```

```python
df_combined = df_TB_1916.join(df_TB_1918, rsuffix='1918')
```

```python
df_combined.loc[:,['Ages','Male Deaths per 10,000 1916', 'Male Deaths per 10,000 1918']].plot(x='Ages',kind='bar')
```
